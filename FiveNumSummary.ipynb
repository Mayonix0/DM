```python
import statistics
import numpy as np
import matplotlib.pyplot as plt

age = [13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70]

mean = statistics.mean(age)
median = statistics.median(age)
mode = statistics.mode(age)

q1 = np.percentile(age, 25)
q2 = np.percentile(age, 50)
q3 = np.percentile(age, 75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = [x for x in age if x < lower_bound or x > upper_bound]

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Q1: {q1}")
print(f"Q2: {q2}")
print(f"Q3: {q3}")
print(f"Interquartile Range (IQR): {iqr}")
print(f"Outliers: {outliers}")

plt.boxplot(age)
plt.title('Box Plot of Age')
plt.show()
```

```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt

file_path = 'day.csv' 
df = pd.read_csv(file_path)

print("Original Dataset:")
print(df)

df = df.drop(['instant','dteday','season','yr','mnth','holiday','weekday','workingday','weathersit'], axis=1)

for column in df.columns:

    q1 = np.percentile(df[column].dropna(), 25)
    q3 = np.percentile(df[column].dropna(), 75)

    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    mean = statistics.mean(df[column].dropna())
    median = statistics.median(df[column].dropna())
    mode = statistics.mode(df[column].dropna())

    print(f"\nStatistics for column '{column}':")
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Number of Outliers: {len(outliers)}")

    plt.boxplot(df[column].dropna())
    plt.title(f'Box Plot of {column} (Including Outliers)')
    plt.show()
```
