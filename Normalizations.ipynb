```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data = {
    'Feature1': [1000, 1005, 2050, 250, 3000],
    'Feature2': [105, 2000, 3000, 400, 505],
    'Feature3': [10000, 2000, 30500, 4000, 5000]
}

file_path = 'test.csv' 
df = pd.read_csv(file_path)

df = pd.DataFrame(data)

plt.figure(figsize=(5, 5))
plt.title('Original Data')
plt.scatter(df.index, df['Feature1'], label='Feature1', marker='o')
plt.scatter(df.index, df['Feature2'], label='Feature2', marker='x')
plt.scatter(df.index, df['Feature3'], label='Feature3', marker='s')
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.legend()
plt.show()

min_max_scaler = MinMaxScaler()
min_max_normalized = min_max_scaler.fit_transform(df)
min_max_df = pd.DataFrame(min_max_normalized, columns=df.columns)

z_score_scaler = StandardScaler()
z_score_normalized = z_score_scaler.fit_transform(df)
z_score_df = pd.DataFrame(z_score_normalized, columns=df.columns)

decimal_normalized = df / 10 ** len(str(df.max().max()))
decimal_df = pd.DataFrame(decimal_normalized, columns=df.columns)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].set_title('Min-Max Normalization')
axs[0].scatter(df.index, min_max_df['Feature1'], label='Feature1', marker='o')
axs[0].scatter(df.index, min_max_df['Feature2'], label='Feature2', marker='x')
axs[0].scatter(df.index, min_max_df['Feature3'], label='Feature3', marker='s')
axs[0].legend()

axs[1].set_title('Z-score Normalization')
axs[1].scatter(df.index, z_score_df['Feature1'], label='Feature1', marker='o')
axs[1].scatter(df.index, z_score_df['Feature2'], label='Feature2', marker='x')
axs[1].scatter(df.index, z_score_df['Feature3'], label='Feature3', marker='s')
axs[1].legend()

axs[2].set_title('Decimal Scaling Normalization')
axs[2].scatter(df.index, decimal_df['Feature1'], label='Feature1', marker='o')
axs[2].scatter(df.index, decimal_df['Feature2'], label='Feature2', marker='x')
axs[2].scatter(df.index, decimal_df['Feature3'], label='Feature3', marker='s')
axs[2].legend()

plt.tight_layout()
plt.show()
```
