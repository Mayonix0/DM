```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importing 3D plotting library
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

file_path = 'Mall_Customers.csv' 
data = pd.read_csv(file_path)
data = data.drop(['CustomerID', 'Gender'], axis=1)

X = data.iloc[:,[0,1,2]].values

kmeans = KMeans(n_clusters=3, init='k-means++')
q = kmeans.fit_predict(X)

cluster_labels = kmeans.labels_

data['Cluster'] = cluster_labels

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for clusters
ax.scatter(X[q==0, 0], X[q==0, 1], X[q==0, 2], c="red", label='Cluster 1')
ax.scatter(X[q==1, 0], X[q==1, 1], X[q==1, 2], c="green", label='Cluster 2')
ax.scatter(X[q==2, 0], X[q==2, 1], X[q==2, 2], c="blue", label='Cluster 3')

# Plot centroids
centroids = kmeans.cluster_centers_
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], marker='*', s=300, c='black', label='Centroids', edgecolors='black')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('K-Means Clustering')

plt.legend()
plt.show()

print("Centroid values:")
for i, centroid in enumerate(centroids):
    print(f"Cluster {i+1} centroid: {centroid}")
```
